```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import itertools

sns.set_theme()
```

```python
part5_heat1 = np.loadtxt('data/4.5-heat1.csv', delimiter=',', skiprows=1)
part5_heat2 = np.loadtxt('data/4.5-heat2.csv', delimiter=',', skiprows=1)
part5_heat3 = np.loadtxt('data/4.5-heat3.csv', delimiter=',', skiprows=1)
part5_cool1 = np.loadtxt('data/4.5-cold1.csv', delimiter=',', skiprows=1)
part5_cool2 = np.loadtxt('data/4.5-cold2.csv', delimiter=',', skiprows=1)
part5_cool3 = np.loadtxt('data/4.5-cold3.csv', delimiter=',', skiprows=1)
```

```python
print(np.max(np.transpose(part5_heat1)[1][275:600]))
print(np.min(np.transpose(part5_heat1)[1][275:600]))

print(np.max(np.transpose(part5_heat2)[1][0:300]))
print(np.min(np.transpose(part5_heat2)[1][0:300]))

print(np.max(np.transpose(part5_heat3)[1][175:500]))
print(np.min(np.transpose(part5_heat3)[1][175:500]))
```

```python
print(np.where(np.transpose(part5_heat1)[1]== -9.304))
print(np.where(np.transpose(part5_heat1)[1]== -10.61))

print(np.where(np.transpose(part5_heat2)[1]== -9.606))
print(np.where(np.transpose(part5_heat2)[1]== -10.309))

print(np.where(np.transpose(part5_heat3)[1]== -9.572))
print(np.where(np.transpose(part5_heat3)[1]== -10.342))
```

```python
print(np.transpose(part5_heat1)[0][367]-np.transpose(part5_heat1)[0][275])
print(np.transpose(part5_heat1)[0][445]-np.transpose(part5_heat1)[0][275])

print(np.transpose(part5_heat2)[0][98]-np.transpose(part5_heat2)[0][0])
print(np.transpose(part5_heat2)[0][163]-np.transpose(part5_heat2)[0][0])

print(np.transpose(part5_heat3)[0][273]-np.transpose(part5_heat3)[0][175])
print(np.transpose(part5_heat3)[0][344]-np.transpose(part5_heat3)[0][175])
```

```python
fig,ax = plt.subplots(figsize = (9,6))
ax.plot(np.transpose(part5_heat1)[0][275:600]-np.transpose(part5_heat1)[0][275],np.transpose(part5_heat1)[1][275:600],'-',markersize = 2,alpha = 1, label = 'with Heating Pack $65.9^\circ C$')
ax.plot(np.transpose(part5_heat2)[0][0:300]-np.transpose(part5_heat2)[0][0],np.transpose(part5_heat2)[1][0:300],'-',markersize = 2,alpha = 1,label = 'with Heating Pack $52.2^\circ C$')
ax.plot(np.transpose(part5_heat3)[0][175:500]-np.transpose(part5_heat3)[0][175],np.transpose(part5_heat3)[1][175:500],'-',markersize = 2,alpha = 1,label = 'with Heating Pack $47.6^\circ C$')

ax.plot(23.799999999999955, -9.304,'.',markersize = 8,color='r')
ax.plot(43.799999999999955, -10.61,'.',markersize = 8,color='r')

ax.plot(25.200000000000045, -9.606,'.',markersize = 8,color='r')
ax.plot(41.59999999999991, -10.309,'.',markersize = 8,color='r')

ax.plot(25.200000000000045, -9.572,'.',markersize = 8,color='r')
ax.plot(43.40000000000009, -10.342,'.',markersize = 8,color='r')

ax.axhline(y = -10.007, color='r', linestyle='--')

ax.set_title('External Temperature Perturbation with Heating Pack',size = 18)
ax.set_xlabel('Time (s)',size = 16)
ax.set_ylabel(r"Temperature $(\circ C)$",size = 16)
ax.tick_params(axis='both',which='major',labelsize=14)
ax.legend(fontsize = 14)

plt.show()

fig.savefig("part5_heating.png", format="png", dpi=1200)
```
